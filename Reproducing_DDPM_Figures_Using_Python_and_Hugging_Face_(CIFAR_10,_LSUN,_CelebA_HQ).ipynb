#### **Reproducing DDPM Figures from the Paper**

This notebook demonstrates to reproduce figures from the *Denoising Diffusion Probabilistic Models* (DDPM) paper using Python.

Due to limitations in Google Colab (especially runtime and GPU resources), I used Hugging Face's `diffusers` library, which provides pretrained DDPM models for datasets such as CIFAR-10, LSUN Bedrooms, LSUN Churches, and CelebA-HQ.

I experimented with different numbers of inference steps and batch sizes to balance image quality and execution time.


### **CelebA-HQ Sample Generation (Alternative using Stable Diffusion)**

The paper uses CelebA-HQ 256×256 for face samples, but the original model setup is quite complex to run on Colab (requires TensorFlow and TPUs). So instead, I used a Stable Diffusion model from Hugging Face (`CompVis/stable-diffusion-v1-4`) to simulate similar results.

I generated portrait-style face images using a descriptive prompt to match CelebA-HQ style.

```python
# Step 1: Install required libraries
!pip install -q diffusers transformers accelerate scipy ftfy
```

```python
# Step 2: Load a pretrained Stable Diffusion model
# (I used this as a substitute for CelebA-HQ due to practical constraints)

from diffusers import StableDiffusionPipeline
import torch

pipe = StableDiffusionPipeline.from_pretrained("CompVis/stable-diffusion-v1-4")
pipe = pipe.to("cuda" if torch.cuda.is_available() else "cpu")
```

```python
# Step 3: Generate 16 portrait-style images
# Prompt is designed to produce clean, high-quality face outputs

prompt = "portrait photo of a young person, high quality, clean background"
images = [pipe(prompt).images[0] for _ in range(16)]
```

```python
# Step 4: Plot the images in a 4x4 grid
import matplotlib.pyplot as plt

fig, axs = plt.subplots(4, 4, figsize=(8, 8))
for i, img in enumerate(images):
    axs[i // 4, i % 4].imshow(img)
    axs[i // 4, i % 4].axis('off')
plt.tight_layout()
plt.show()
```

## **CIFAR-10 Sample Generation (32×32 Images)**

For this part, I used the pretrained DDPM model from Hugging Face trained on the CIFAR-10 dataset. The image resolution is 32×32, which is much lower than LSUN or CelebA-HQ, but this is consistent with the original DDPM paper.

Initially, I used 1000 denoising steps to stay close to the paper setup. The image generation was quite slow in Colab, but still manageable for 16 samples. I also added GPU memory cleanup between generations to avoid crashes.

```python
!pip install torch torchvision diffusers transformers
```

```python
# Step 1: Load the pretrained DDPM model for CIFAR-10 (32x32 images)

from diffusers import DDPMPipeline
import torch

# Load pretrained DDPM model from Hugging Face (trained on CIFAR-10)
pipeline = DDPMPipeline.from_pretrained("google/ddpm-cifar10-32")
pipeline.to("cuda" if torch.cuda.is_available() else "cpu")
```

```python
# Step 2: Generate 16 samples using 1000 inference steps
#(Note: 1000 steps = high quality but slow in Colab)

import torch
import matplotlib.pyplot as plt
import gc

images = []

for _ in range(16):
    out = pipeline(num_inference_steps=1000).images[0]
    images.append(out)

    #Clear GPU memory to avoid OOM error (especially important in Colab)
    del out
    torch.cuda.empty_cache()
    gc.collect()

#Step 3: Plot the 16 images in a 4x4 grid
fig, axs = plt.subplots(4, 4, figsize=(8, 8))
for i, img in enumerate(images):
    axs[i // 4, i % 4].imshow(img)
    axs[i // 4, i % 4].axis('off')
plt.tight_layout()
plt.show()
```

#### **Notes**

The model generated diverse and recognizable low-res images from CIFAR-10 categories (e.g., cars, animals, planes).  
Although the resolution is small, this is expected since CIFAR-10 is a 32×32 dataset.

If needed, I could reduce the number of inference steps (e.g., to 250) to speed up generation, but I wanted to keep this one at 1000 steps to stay as close to the original paper as possible.


### **LSUN Bedroom Image Generation**

In the DDPM paper, LSUN Bedroom was used to demonstrate high-resolution image generation (256×256). I used Hugging Face’s pretrained model `google/ddpm-bedroom-256` for this part.

I started with 1000 inference steps, but it was too slow in Colab — even one image took almost an hour. So I reduced the steps to 250 to balance image quality and runtime.

Below are 4 samples generated using the DDPM pipeline. Each image was generated one by one, which helped avoid GPU memory overload.


```python
# Step 1: Load the pretrained DDPM model for LSUN Bedroom (256x256 resolution)
from diffusers import DDPMPipeline
import torch
import matplotlib.pyplot as plt

pipeline = DDPMPipeline.from_pretrained("google/ddpm-bedroom-256")
pipeline.to("cuda" if torch.cuda.is_available() else "cpu")

# Step 2: Generate 4 bedroom images using 250 inference steps (faster than 1000)
images = [pipeline(num_inference_steps=250).images[0] for _ in range(4)]

# Step 3: Plot the images in a 2x2 grid
fig, axs = plt.subplots(2, 2, figsize=(6, 6))
for i, img in enumerate(images):
    axs[i // 2, i % 2].imshow(img)
    axs[i // 2, i % 2].axis('off')
plt.tight_layout()
plt.suptitle("LSUN Bedroom Samples (250 Steps)", fontsize=14)
plt.show()
```

### **LSUN Church Image Generation**

Similar to LSUN Bedroom, I used the pretrained DDPM model `google/ddpm-church-256` to generate church images at 256×256 resolution.

Again, I chose 250 inference steps instead of 1000 to reduce total generation time and avoid GPU timeouts in Colab. The results still look reasonably detailed and capture the architectural patterns typical of LSUN Church images.

Below are 4 generated samples.

```python
#Install required libraries silently (in case not installed earlier)
!pip install diffusers transformers torch torchvision --quiet
```

```python
#Step 1: Load pretrained DDPM model for LSUN Church (256x256 resolution)
from diffusers import DDPMPipeline
import torch
import matplotlib.pyplot as plt

pipeline = DDPMPipeline.from_pretrained("google/ddpm-church-256")
pipeline.to("cuda" if torch.cuda.is_available() else "cpu")

# Step 2: Generate 4 church images with 250 denoising steps
images = [pipeline(num_inference_steps=250).images[0] for _ in range(4)]

# Step 3: Plot in a 2x2 grid
fig, axs = plt.subplots(2, 2, figsize=(6, 6))
for i, img in enumerate(images):
    axs[i // 2, i % 2].imshow(img)
    axs[i // 2, i % 2].axis('off')
plt.tight_layout()
plt.suptitle("LSUN Church Samples (DDPM)", fontsize=16)
plt.show()
```
